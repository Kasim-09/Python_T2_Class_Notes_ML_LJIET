# SVM (Support Vector Machine)
- kernel, hyperplane, outlier, margin distance formula
- more less C ---> Error ignorance less (So it trains model more and more and at last it overfit)
- C(regularization parameter)

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix,classification_report
from sklearn.svm import SVC
````

````python
f = pd.read_csv("Iris.csv")
df = pd.DataFrame(f)
df
````

````python
x = df.drop(["Id","Species"],axis=1)
y = df["Species"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model = SVC(kernel="rbf",C=1,random_state=42) # rbf->Radial Basis Function and svc-> support vector classifier
model.fit(x_train,y_train)
pre = model.predict(x_test)
acc = accuracy_score(y_test,pre)
print(acc)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(
    x=y_test,
    y=pre
)
````
